```python
country_gas_levels = aqm.groupby('Country').sum()

# Find the country with the highest levels for each gas
highest_co = country_gas_levels['CO AQI Value'].idxmax()
highest_ozone = country_gas_levels['Ozone AQI Value'].idxmax()
highest_no2 = country_gas_levels['NO2 AQI Value'].idxmax()
highest_pm25 = country_gas_levels['PM2.5 AQI Value'].idxmax()

print("Country with the highest CO levels:", highest_co)
print("Country with the highest Ozone levels:", highest_ozone)
print("Country with the highest NO₂ levels:", highest_no2)
print("Country with the highest PM2.5 levels:", highest_pm25)

# Get the values for visualization
values = [
    country_gas_levels.loc[highest_co, 'CO AQI Value'],
    country_gas_levels.loc[highest_ozone, 'Ozone AQI Value'],
    country_gas_levels.loc[highest_no2, 'NO2 AQI Value'],
    country_gas_levels.loc[highest_pm25, 'PM2.5 AQI Value']
]

# Prepare labels
countries = [
    f'CO - {highest_co}',
    f'Ozone - {highest_ozone}',
    f'NO₂ - {highest_no2}',
    f'PM2.5 - {highest_pm25}'
]

# Plotting the pie chart
plt.figure(figsize=(7, 7))
plt.pie(values, labels=countries, autopct='%1.1f%%',
        colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'], startangle=140)
plt.title('Highest Pollutant Levels by Country (2024)')
plt.show()
```

```python
# Group by country and sum AQI values for pollutants
country_sums = aqm.groupby('Country')[['CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']].sum()

# Find the country with the highest AQI for each pollutant
highest_co_country = country_sums['CO AQI Value'].idxmax()
highest_ozone_country = country_sums['Ozone AQI Value'].idxmax()
highest_no2_country = country_sums['NO2 AQI Value'].idxmax()
highest_pm25_country = country_sums['PM2.5 AQI Value'].idxmax()

# Get the AQI values for these countries
values = [
    country_sums.loc[highest_co_country, 'CO AQI Value'],
    country_sums.loc[highest_ozone_country, 'Ozone AQI Value'],
    country_sums.loc[highest_no2_country, 'NO2 AQI Value'],
    country_sums.loc[highest_pm25_country, 'PM2.5 AQI Value']
]

# Labels combining pollutant and country
labels = [
    f'CO - {highest_co_country}',
    f'Ozone - {highest_ozone_country}',
    f'NO2 - {highest_no2_country}',
    f'PM2.5 - {highest_pm25_country}'
]

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(labels, values, color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
plt.ylabel('AQI Value')
plt.title('Highest Pollutant Levels by Country (2024)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The chart highlights the countries with the highest total AQI values for each major pollutant. Each bar represents a pollutant-country pair where pollution levels are most severe. This reveals how different countries face distinct pollution challenges.

```python

# Set the style for better visuals
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

# 2️. AQI Values by Top 10 Countries (Aggregated)
top_countries = aqm.groupby('Country')['AQI Value'].mean().nlargest(10).reset_index()

# Fix: Use hue and remove the legend
sns.barplot(x='AQI Value', y='Country', data=top_countries, hue='Country', dodge=False, palette='plasma')
plt.title('Top 10 Countries by Average AQI Value')
plt.legend([], [], frameon=False)  # Remove the redundant legend
plt.show()
```

The bar chart highlights the **top 10 countries with the highest average AQI values**, reflecting the areas with the poorest overall air quality in the dataset. The differences in bar lengths emphasize the variation in pollution severity among these countries, with some countries exhibiting significantly higher average AQI levels than others. This visualization helps identify key regions where air pollution mitigation efforts are most urgently needed.

```python

# Set the style
sns.set_style("whitegrid")
plt.figure(figsize=(14, 6))

# List of pollutant columns
pollutants = ['CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']

# Prepare data for plotting
pollutant_data = pd.DataFrame({
    'Pollutant': pollutants,
    'Average AQI Value': [aqm[col].mean() for col in pollutants]
})

# Plot using hue directly
sns.barplot(data=pollutant_data, x='Pollutant', y='Average AQI Value', hue='Pollutant', dodge=False, palette='magma')
plt.title('Average AQI Values by Pollutant Type')
plt.legend([], [], frameon=False)  # Remove the extra legend
plt.show()
```

The bar chart reveals that among the major pollutants, **PM2.5 has the highest average AQI value**, indicating that fine particulate matter is the most significant contributor to air pollution in the dataset. NO2 also shows a relatively high average, reflecting its impact on urban air quality. In contrast, **CO and Ozone have lower average AQI values**, suggesting their pollution levels are generally less severe across the sampled cities.

```python

# Set the style
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

# Group by Country and AQI Category, then get the average AQI
country_aqi = aqm.groupby(['Country', 'AQI Category'])['AQI Value'].mean().reset_index()

# Sort by AQI Value and pick the Top 10 countries
top_countries = country_aqi.sort_values(by='AQI Value', ascending=False).head(10)

# Plot the bar chart with hue as AQI Category
sns.barplot(x='AQI Value', y='Country', hue='AQI Category', data=top_countries, palette='Set3')
plt.title('Top 10 Countries by Average AQI Value Categorized by AQI Category')
plt.legend(title='AQI Category')
plt.show()
```

The chart displays the top 10 countries with the highest average AQI values, categorized by their AQI levels. It shows how different AQI categories contribute to the overall pollution in each country, with many countries having a mix of moderate to unhealthy air quality segments. This visualization highlights that even among the most polluted countries, air quality can vary significantly across regions and pollution severity levels.

```python


# Pick two categories for side-by-side comparison (e.g., 'Good' and 'Unhealthy')
categories = ['Good', 'Unhealthy']

# Extract AQI Values for those categories
cat1_aqi = aqm[aqm['AQI Category'] == categories[0]]['AQI Value']
cat2_aqi = aqm[aqm['AQI Category'] == categories[1]]['AQI Value']

# Define bins
aqi_bins = range(0, 501, 25)

# Create side-by-side histograms
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

sns.histplot(cat1_aqi, bins=aqi_bins, color='green', ax=ax1, edgecolor='black')
ax1.set_title(f'{categories[0]} AQI Values')
ax1.set_xlabel('AQI Value')
ax1.set_ylabel('Count')

sns.histplot(cat2_aqi, bins=aqi_bins, color='red', ax=ax2, edgecolor='black')
ax2.set_title(f'{categories[1]} AQI Values')
ax2.set_xlabel('AQI Value')

plt.tight_layout()
plt.show()
```

The histograms clearly show that cities in the **“Good”** AQI category have AQI values tightly clustered at the lower end of the scale, mostly below 50, indicating healthy air quality. In contrast, the “Unhealthy” category displays a wider spread of much higher AQI values, often exceeding **150**, reflecting severe air pollution in those cities. This side-by-side comparison highlights the stark difference in air quality between healthier and more polluted environments.

```python

# Calculate average CO AQI Value per country
country_avg_co = aqm.groupby('Country')['CO AQI Value'].mean().reset_index()

# Top 3 countries with highest average CO AQI Value
top_countries = country_avg_co.nlargest(5, 'CO AQI Value')['Country'].tolist()
top_data = aqm[aqm['Country'].isin(top_countries)]

# Bottom 3 countries with lowest average CO AQI Value
bottom_countries = country_avg_co.nsmallest(5, 'CO AQI Value')['Country'].tolist()
bottom_data = aqm[aqm['Country'].isin(bottom_countries)]

# Plot for highest CO AQI countries
fig_top = px.box(top_data, x='Country', y='CO AQI Value', color='Country',
                title='Top 5 Countries by Average CO AQI Value')
fig_top.show()

# Plot for lowest CO AQI countries
fig_bottom = px.box(bottom_data, x='Country', y='CO AQI Value', color='Country',
                    title='Bottom 5 Countries by Average CO AQI Value')
fig_bottom.show()
```

```python

# Calculate average Ozone AQI Value per country
country_avg_o3 = aqm.groupby('Country')['Ozone AQI Value'].mean().reset_index()

# Top 3 countries with highest average Ozone AQI Value
top_countries = country_avg_o3.nlargest(3, 'Ozone AQI Value')['Country'].tolist()
top_data = aqm[aqm['Country'].isin(top_countries)]

# Bottom 3 countries with lowest average Ozone AQI Value
bottom_countries = country_avg_o3.nsmallest(3, 'Ozone AQI Value')['Country'].tolist()
bottom_data = aqm[aqm['Country'].isin(bottom_countries)]

# Plot for highest Ozone AQI countries
fig_top = px.box(top_data, x='Country', y='Ozone AQI Value', color='Country',
                title='Top 3 Countries by Average Ozone AQI Value')
fig_top.show()

# Plot for lowest Ozone AQI countries
fig_bottom = px.box(bottom_data, x='Country', y='Ozone AQI Value', color='Country',
                    title='Bottom 3 Countries by Average Ozone AQI Value')
fig_bottom.show()
```

```python

# Calculate average NO2 AQI Value per country
country_avg_o3 = aqm.groupby('Country')['NO2 AQI Value'].mean().reset_index()

# Top 3 countries with highest average NO2 AQI Value
top_countries = country_avg_o3.nlargest(3, 'NO2 AQI Value')['Country'].tolist()
top_data = aqm[aqm['Country'].isin(top_countries)]

# Bottom 3 countries with lowest average NO2 AQI Value
bottom_countries = country_avg_o3.nsmallest(3, 'NO2 AQI Value')['Country'].tolist()
bottom_data = aqm[aqm['Country'].isin(bottom_countries)]

# Plot for highest NO2 AQI countries
fig_top = px.box(top_data, x='Country', y='NO2 AQI Value', color='Country',
                title='Top 3 Countries by Average NO2 AQI Value')
fig_top.show()

# Plot for lowest NO2 AQI countries
fig_bottom = px.box(bottom_data, x='Country', y='NO2 AQI Value', color='Country',
                    title='Bottom 3 Countries by Average NO2 AQI Value')
fig_bottom.show()
```

```python

# Calculate average PM2.5 AQI Value per country
country_avg_o3 = aqm.groupby('Country')['PM2.5 AQI Value'].mean().reset_index()

# Top 3 countries with highest average PM2.5 AQI Value
top_countries = country_avg_o3.nlargest(3, 'PM2.5 AQI Value')['Country'].tolist()
top_data = aqm[aqm['Country'].isin(top_countries)]

# Bottom 3 countries with lowest average PM2.5 AQI Value
bottom_countries = country_avg_o3.nsmallest(3, 'PM2.5 AQI Value')['Country'].tolist()
bottom_data = aqm[aqm['Country'].isin(bottom_countries)]

# Plot for highest PM2.5 AQI countries
fig_top = px.box(top_data, x='Country', y='PM2.5 AQI Value', color='Country',
                title='Top 3 Countries by Average PM2.5 AQI Value')
fig_top.show()

# Plot for lowest PM2.5 AQI countries
fig_bottom = px.box(bottom_data, x='Country', y='PM2.5 AQI Value', color='Country',
                    title='Bottom 3 Countries by Average PM2.5 AQI Value')
fig_bottom.show()
```

The plot shows that the overall AQI value generally correlates with pollutant levels, especially **PM2.5 and NO2**, which tend to have higher values in the most polluted cities. Ozone and CO levels vary more independently, sometimes peaking where **overall AQI is moderate**. This indicates that while particulate matter and nitrogen dioxide are major contributors to poor air quality, ozone and carbon monoxide pollution depend on different environmental factors and sources across cities.


```python

plt.figure(figsize=(12, 6))

# Bar Plot for AQI Value
sns.barplot(x='City', y='AQI Value', data=aqm.head(10), color='skyblue', label='AQI Value')

# Line Plot for Ozone AQI Value
sns.lineplot(x='City', y='Ozone AQI Value', data=aqm.head(10), color='orange', marker='o', label='Ozone AQI Value')

plt.title('AQI Value vs Ozone AQI Value for Top 10 Cities')
plt.xlabel('City')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# Select Top 10 Cities based on AQI Value
top_10_cities = aqm.nlargest(10, 'AQI Value')

# Set figure size
plt.figure(figsize=(14, 7))

# Plot the lines
sns.lineplot(x='City', y='AQI Value', data=top_10_cities, marker='o', label='AQI Value', color='blue')
sns.lineplot(x='City', y='Ozone AQI Value', data=top_10_cities, marker='o', label='Ozone AQI Value', color='orange')
sns.lineplot(x='City', y='CO AQI Value', data=top_10_cities, marker='o', label='CO AQI Value', color='green')
sns.lineplot(x='City', y='NO2 AQI Value', data=top_10_cities, marker='o', label='NO2 AQI Value', color='red')
sns.lineplot(x='City', y='PM2.5 AQI Value', data=top_10_cities, marker='o', label='PM2.5 AQI Value', color='purple')

# Customize the plot
plt.title('Pollutant Levels vs AQI Value for Top 10 Cities')
plt.xlabel('City')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

The above line graph represents AQI value and PM2.5 are closely related, showing that particulate matter(PM2.5) is a key factor in pollution. Ozone levels sometimes peak even when AQI is moderate, meaning ozone pollution depends on other factors. Cities with high PM2.5 usually have high AQI, but ozone pollution can behave differently across cities.

```python
# List of pollutants to visualize
pollutants = ['AQI Value', 'Ozone AQI Value', 'CO AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value']

plt.figure(figsize=(15, 8))

# Plot each pollutant's distribution
for i, pollutant in enumerate(pollutants, 1):
    plt.subplot(2, 3, i)
    sns.histplot(aqm[pollutant], bins=20, kde=True, color='teal', edgecolor='black')
    plt.title(f'Distribution of {pollutant}')
    plt.xlabel(pollutant)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

The graph shows most cities have pollutant levels clustered within the lower range (0–50 AQI), indicating generally good air quality. However, **PM2.5 and NO2** have wider spreads, with values extending beyond 100 AQI, reflecting unhealthy pollution levels in some cities. Ozone and CO mostly stay within moderate ranges, showing less extreme variation across locations.

**Overall Overview**

The analysis of the air pollution dataset reveals significant insights into global air quality patterns. The average AQI values indicate that **PM2.5 and NO2 are the dominant contributors** to poor air quality across most polluted cities, while **CO and Ozone levels vary independently**, suggesting different environmental factors affecting their concentrations.

Visualizations of the top 10 most polluted countries and cities highlight regions with critical air pollution challenges, with some countries consistently ranking** high across multiple pollutants**. The scatter plots show a positive correlation between overall AQI and individual pollutants like Ozone and PM2.5, reflecting their impact on general air quality.

Additionally, the category based analysis (Good vs Unhealthy) underlines the disparity between regions with healthy air and those struggling with severe pollution, while the highest pollutant levels by country point out specific nations leading in emissions for CO, Ozone, NO2, and PM2.5.

The Correlation graph represents the datasets consisting of **positive, negative and neutral traversal of gases** in the atmosphere. This helps in identifing the highly affecting gases in atmosphere.

Overall, the dataset underscores the need for **targeted air quality management** in heavily impacted regions and stronger regulations to reduce particulate matter and nitrogen dioxide, which are the largest threats to global air quality. This can also be handled for maintaining the atmosphere level inplace of highly affecting gases through **physical or biological activities**.
